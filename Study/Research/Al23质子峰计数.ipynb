但对于弱峰，这种扣除方法无效。因此决定采用最简单的线性本底估计的方法来扣除本底。

```c++
d2e3->Draw();
d2e3->ShowPeaks(2,"",0.002);
c1->Draw();
```

```c++
map<int,int> range;
map<int,int>::iterator iter;
```

```c++
range[44]=55;
range[69]=78;
range[100]=107;
range[120]=124;
range[126]=144;
range[171]=180;
range[254]=270;
range[282]=288;
range[300]=312;
range[397]=408;
range[440]=457;
range[565]=575;
range[584]=596;
range[606]=615;
range[677]=684;
range[341]=359;
range[248]=253;
```

```c++
for(iter = range.begin(); iter != range.end(); iter++)  
{
    cout<<iter->first<<'\t'<<iter->second<<'\n';
    double peak = 10 * d2e3->Integral(iter->first,iter->second);
    double bkg = 0.5*(d2e3->GetBinCenter(iter->second)
                      -d2e3->GetBinCenter(iter->first))
                 *(d2e3->GetBinContent(iter->second)
                  +d2e3->GetBinContent(iter->first));
    double count=peak-bkg;
    cout<<"peak: "<<peak<<'\t'<<"bkg: "<<bkg<<'\t'<<"count: "<<count<<'\n';
}
    
```
